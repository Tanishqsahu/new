```python
from sklearn.preprocessing import StandardScaler, RobustScaler
import pandas as pd

# 1. Define the features for each scaler
# We'll assume the original data has 'Time' and 'Amount' (and V1-V28, which we skip)
time_col = 'Time'
amount_col = 'Amount'

# 2. Initialize the Scalers
scaler_standard = StandardScaler()
scaler_robust = RobustScaler()

# 3. Apply Standard Scaler to 'Time'
df[time_col + '_scaled'] = scaler_standard.fit_transform(df[[time_col]])

# 4. Apply Robust Scaler to 'Amount'
df[amount_col + '_scaled'] = scaler_robust.fit_transform(df[[amount_col]])

# 5. Drop the original unscaled columns
df = df.drop(columns=[time_col, amount_col])

# The DataFrame is now clean and ready for modeling!
```

Now training model

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import recall_score, precision_score, f1_score, roc_auc_score, confusion_matrix
import pandas as pd

# Assume 'df' is your DataFrame with V-features, 'log_amount', 'hour_sin', 'hour_cos', and 'class'

# 1. Define Features (X) and Target (y)
X = df.drop('Class', axis=1)
y = df['Class']

# 2. Scale the Non-PCA Features
# Only 'Time', 'Amount', and their engineered variants need scaling, as V1-V28 are already scaled by PCA.
scaler = StandardScaler()
X[['Time_scaled', 'Amount_scaled', 'log_amount']] = scaler.fit_transform(X[['Time_scaled', 'Amount_scaled', 'log_amount']])

# Drop the original skewed 'Amount' column if 'log_amount' is present, and optionally 'Time'
X = X.drop('Amount_scaled', axis=1) 
# Note: You can also keep 'Amount' and 'Time' if you scaled them; the model will decide their importance.
```

```python
# Calculate the imbalance ratio (Total Normal / Total Fraud)
# This value tells XGBoost how much more important the minority class is.
neg, pos = np.bincount(y)
scale_pos_weight = neg / pos

print(f"Normal Count: {neg}, Fraud Count: {pos}")
print(f"Scale Position Weight: {scale_pos_weight:.2f}") 
# Expected value is around 577.
```

```python
# Initialize Stratified K-Fold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Lists to store metrics from each fold
recall_scores = []
precision_scores = []
f1_scores = []
auc_scores = []

# Loop through each fold
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Initialize XGBoost Classifier
    # Use the calculated scale_pos_weight to address imbalance
    xgb_model = XGBClassifier(
        objective='binary:logistic',
        n_estimators=100,
        learning_rate=0.1,
        max_depth=5,  # Moderate depth is usually good
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        scale_pos_weight=scale_pos_weight, # <--- THE KEY IMBALANCE HANDLING PARAMETER
        use_label_encoder=False,
        eval_metric='logloss'
    )

    # Train the model
    xgb_model.fit(X_train, y_train)

    # Predict probabilities (for AUC) and classes (for Recall/Precision)
    y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]
    y_pred = xgb_model.predict(X_test)

    # Evaluate metrics
    recall_scores.append(recall_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))
    auc_scores.append(roc_auc_score(y_test, y_pred_proba))
```

```python
print("--- Cross-Validation Results ---")
print(f"Avg. Recall:   {np.mean(recall_scores):.4f} (Crucial: minimizes False Negatives)")
print(f"Avg. Precision: {np.mean(precision_scores):.4f} (Minimizes False Positives)")
print(f"Avg. F1-Score: {np.mean(f1_scores):.4f} (Balance between Precision & Recall)")
print(f"Avg. ROC-AUC:  {np.mean(auc_scores):.4f} (Model's discrimination ability)")
```
